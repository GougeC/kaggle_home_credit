```python
bureau_features = bureau[['sk_id_curr','amt_credit_sum','amt_credit_sum_debt']].groupby('sk_id_curr').sum()
bureau_features.head()
```

```python
bureau_features.reset_index(inplace = True)
```

```python
bureau_features.head()
```

```python
credit_types = bureau['credit_type'].unique()
bureauids = bureau.sk_id_curr.unique()
```

```python
def get_credit_types(sk_id):
    if sk_id in bureauids:
        types = []
        prevs = bureau[bureau['sk_id_curr'] == sk_id]
        for t in credit_types:
            if np.any(prevs['credit_type'] == t):
                types.append(t)
        return types
    else:
        return []
```

```python
bureau_features['credit_types'] = bureau_features['sk_id_curr'].apply(get_credit_types)
for t in credit_types:
    bureau_features['had_'+t] = bureau_features['credit_types'].apply(lambda x: 1 if t in x else 0)
```

```python
prev = pd.read_csv('data/previous_application.csv')
prev.rename(columns={x:x.lower() for x in prev.columns}, inplace=True)
```

```python
prev.head()
```

```python
categorical_columns = [x for x in prev.columns if prev[x].dtype == object]
```

```python
uniques = {}
for x in categorical_columns:
    uniques[x] = prev[x].unique()
    
```

```python
import multiprocessing
from multiprocessing import Pool
NUM_CPU = multiprocessing.cpu_count()
print(NUM_CPU)
```

```python
def get_id_dict(args):
    skid,uniques,df,categorical_columns = args
    prevdict = {}
    for c in categorical_columns:
        prevdict[c] = []
        for y in uniques[c]:
            if any(df[c] == y):
                prevdict[c].append(c)
    return (skid,prevdict)
```

```python
dfdict = {}
ids = prev['sk_id_curr'].unique()

for x in ids:
    dfdict[x] =  prev[prev['sk_id_curr'] == x]
    
inputs = []
for skid in ids:
    if skid in dfdict:
        inputs.append((skid,uniques,dfdict[skid],categorical_columns))
pool = Pool(NUM_CPU)
```

```python
results = pool.map(get_id_dict,inputs)
```
