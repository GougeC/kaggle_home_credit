```python
bureau_features = bureau[['sk_id_curr','amt_credit_sum','amt_credit_sum_debt']].groupby('sk_id_curr').sum()
bureau_features.head()
```

```python
bureau_features.reset_index(inplace = True)
```

```python
bureau_features.head()
```

```python
credit_types = bureau['credit_type'].unique()
bureauids = bureau.sk_id_curr.unique()
```

```python
def get_credit_types(sk_id):
    if sk_id in bureauids:
        types = []
        prevs = bureau[bureau['sk_id_curr'] == sk_id]
        for t in credit_types:
            if np.any(prevs['credit_type'] == t):
                types.append(t)
        return types
    else:
        return []
```

```python
bureau_features['credit_types'] = bureau_features['sk_id_curr'].apply(get_credit_types)
for t in credit_types:
    bureau_features['had_'+t] = bureau_features['credit_types'].apply(lambda x: 1 if t in x else 0)
```
